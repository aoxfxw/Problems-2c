 #  5 - Pre-process dataset

```python
# to efficiently analyse the dataset, it must be pre-processed.
```

```python
news
```

```python
# move column 'headline' into a list
news_list = news['headline'].tolist() 
```

```python
news_list
```

```python
#tokenize
preprocess = []
for i in news_list:
    words = word_tokenize(i)
    preprocess.append(words)
```

```python
preprocess
```

```python
# remove punctuation
for i in preprocess:
    for index_j, j in enumerate(i):
        if j in punct:
            i.pop(index_j)
```

```python
preprocess
```

```python
#lower all cases
preprocess = [[j.lower() for j in i] for i in preprocess]
```

```python
preprocess
```

```python
# move column 'topic' into a list
topics = news['topic'].tolist() 
```

```python
topics
```

```python
#tokenize
topic = []
for i in topics:
    words = word_tokenize(i)
    topic.append(words)
```

```python
# remove punctuation
for i in topic:
    for index_j, j in enumerate(i):
        if j in punct:
            i.pop(index_j)
```

```python
#lower all cases
topics = [[j.lower() for j in i] for i in topic]
```

```python
topics
```

```python
#export to dataframe
news = pd.DataFrame()

news['headline'] = preprocess
news['topic'] = topics
```
