 #  5 - Pre-process dataset

```python
# to efficiently analyse the dataset, it must be pre-processed.
```

```python
govt
```

```python
# move column 'text' into a list
govt_list = govt['text'].tolist() 
```

```python
govt_list
```

```python
#tokenize
preprocess = []
for i in govt_list:
    words = word_tokenize(i)
    preprocess.append(words)
```

```python
preprocess
```

```python
# remove punctuation
for i in preprocess:
    for index_j, j in enumerate(i):
        if j in punct:
            i.pop(index_j)
```

```python
preprocess
```

```python
#lower all cases
preprocess = [[j.lower() for j in i] for i in preprocess]
```

```python
preprocess
```

```python
#export to dataframe
govt = pd.DataFrame()

govt['text'] = preprocess
```
